```python
def add(a, b):
    return a + b
```

```python
def subtract(a, b):
    return a - b
```

```python
def multiply(a, b):
    return a * b
```

```python
def devide(a, b):
    if b == 0:
        return 'error'
    return a / b
```

```python
def is_number(n):
    if type(n) == int or type(n) == float:
        return True
    return False
```

```python
def is_basic_operator(o):
    if o in ['+', '-', '*', '/']:
        return True
    return False
```

```python
def basic_calc(a, o, b):
    if not is_number(a) or not is_number(b):
        return 'error'
    
    if not is_basic_operator(o):
        return 'error'
    
    if o == '+':
        return add(a, b)
    
    if o == '-':
        return subtract(a, b)
    
    if o == '*':
        return multiply(a, b)
    
    if o == '/':
        return devide(a, b)
```

```python
basic_calc(1, '+', 2)
```

```python
basic_calc(2, '-', 1)
```

```python
basic_calc(1, '*', 2)
```

```python
basic_calc(2, '/', 1)
```
