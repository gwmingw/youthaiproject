# 🤖 [AI 해커톤 사전 과제] LLM 파트너와 함께하는 디자인 씽킹 워크북

안녕하세요, 예비 해커톤 참가자 여러분! 
이 워크북은 여러분이 대형 언어 모델(LLM)과 대화하며 일상의 불편함으로부터 멋진 **AI 서비스 아이디어**를 도출할 수 있도록 돕는 **'아이디어 네비게이터'**입니다.

### 💡 디자인 씽킹(Design Thinking)이란?
사용자의 불편함에 깊이 공감하고, 진짜 문제를 정의한 뒤, 창의적인 해결책을 빠르게 만들어 테스트하는 인간 중심의 혁신 방법론입니다. 이번 해커톤에서는 특별히 **AI 코치**와 함께 이 과정을 밟아볼 거예요!

### 🛠️ 사용 방법
1. 각 단계별로 마련된 **[Python 코드 셀]**을 실행하여 프롬프트(명령어)를 확인합니다.
2. 프롬프트 내용 중 `< >`로 되어 있는 빈칸(나의 관심사, 불편함 등)을 여러분의 이야기로 채워 넣으세요.
3. 완성된 프롬프트를 복사해서 **ChatGPT, Gemini** 등의 생성형 AI 서비스에 입력하고 대화를 나누세요.
4. AI가 주는 답변을 바탕으로 생각을 발전시켜 나가며 마지막 단계까지 완주해 보세요!

---
## 📌 [Step 1] 공감하기(Empathize) & 문제 정의하기(Define)

좋은 아이디어는 주변에 대한 세심한 관찰에서 시작됩니다. 내가 학교, 일상, 취미 생활 등에서 느꼈던 작은 불편함이나 아쉬움을 떠올려 보세요. 

**👉 하단의 코드에서 괄호 `< >` 안을 나만의 내용으로 수정한 뒤, 코드 셀을 실행하여 생성된 프롬프트로 AI에게 질문하세요.**

```python
# 1단계 프롬프트 생성 코드
interest_area = "<학교 생활 및 공부 관리>" # 예: 학교 생활, 환경, 진로, 취미 등
pain_point = "<수행평가 기간만 되면 여러 과목의 마감일이 겹쳐서 계획을 세우기가 너무 힘들고 스트레스를 받아.>" # 내가 느끼는 구체적인 불편함

prompt_step1 = f"""
너는 청소년 AI 해커톤의 친절하고 전문적인 디자인 씽킹 코치야. 
내가 일상에서 겪는 불편함이나 관심사를 바탕으로 사용자의 마음에 공감하고 '진짜 문제'를 정의할 수 있도록 도와줘.

[나의 정보]
- 내가 관심 있는 분야: {interest_area}
- 내가 또는 주변 친구들이 겪는 구체적인 불편함: {pain_point}

[코치의 역할 및 요구사항]
1. 내가 입력한 불편함을 분석해서, 사용자가 이 상황에서 느낄 심리적 상태와 진짜 니즈(Needs)가 무엇인지 '사용자 관점'에서 깊이 공감해 줘.
2. 이 불편함을 가장 격렬하게 겪고 있을 법한 구체적인 '가상 페르소나(가상의 타겟 사용자)'를 한 명 입체적으로 만들어줘. (이름, 나이, 처한 상황, 성격 특징 포함)
3. 이 페르소나를 돕기 위한 "우리가 어떻게 하면 ~할 수 있을까?(How Might We)" 형태의 핵심 문제 정의 문장을 3가지 방향성으로 다채롭게 추천해 줘.
4. 답변의 마지막에는 나에게 '3가지 문제 정의 문장 중 어떤 것이 가장 마음에 드는지, 혹은 수정하고 싶은지' 다정하게 질문을 던져줘.
"""

print(prompt_step1.strip())
```

---
## 💡 [Step 2] AI 아이디어 발상하기(Ideate)

문제가 명확해졌다면, 이제 AI 기술을 접목할 차례입니다! 
해커톤이라는 한정된 시간 안에 청소년 팀이 직접 구현할 수 있으면서도 실용적인 **AI 서비스/챗봇 아이디어**를 AI와 함께 브레인스토밍해 봅니다.

**👉 AI와의 대화를 통해 결정된 '최종 문제 정의 문장'을 아래 코드에 넣고 실행하세요.**

```python
# 2단계 프롬프트 생성 코드
chosen_hmw = "<우리가 어떻게 하면 마감일이 겹친 학생들이 불안해하지 않고, 우선순위에 맞춰 자동으로 실천 가능한 계획을 제안받을 수 있을까?>" # 1단계에서 AI가 추천해 준 문장 중 선택

prompt_step2 = f"""
우리가 함께 정의한 핵심 문제는 다음과 같아:
"{chosen_hmw}"

너는 이제 창의적인 AI 서비스 기획자이자 기술 전문가야. 
이 문제를 'AI 에이전트'나 'LLM(대형 언어 모델) 기반 서비스'로 해결할 수 있는 독창적인 아이디어를 브레인스토밍해 주려고 해. 
청소년들이 해커톤 기간(단기 프로젝트) 동안 프롬프트 엔지니어링이나 노코드 툴, 혹은 간단한 웹/앱 API 연동으로 실제로 프로토타입을 만들어볼 수 있는 수준이어야 해.

[요구사항]
1. 서로 다른 접근 방식을 가진 독창적인 AI 서비스 아이디어 3가지를 제안해 줘.
2. 각 아이디어는 반드시 다음 항목을 포함해야 해:
   - 서비스 이름 (기억에 남는 가제)
   - 핵심 AI 기능 (AI가 유저와 대화하거나 데이터를 처리할 때 어떤 스마트한 역할을 하는지)
   - 기술적 실현 가능성 (청소년 수준에서 왜 구현하기 수월한지, 예: '이건 ChatGPT API에 페르소나 프롬프트를 부여하는 것만으로 구현 가능해!' 등)
3. 3가지 아이디어 중 청소년 해커톤 목적에 가장 부합하고, 구현 대비 유저의 만족도가 높을 것 같은 '원픽(One-Pick) 추천 아이디어'와 그 이유를 설명해 줘.
4. 마지막에는 나에게 '이 아이디어들 중 어떤 방향이 가장 흥미로운지, 혹은 결합하고 싶은 기능이 있는지' 질문해 줘.
"""

print(prompt_step2.strip())
```

---
## 🧪 [Step 3] 가상 프로토타이핑 & 페르소나 테스트(Prototype & Test)

실제로 코딩을 시작하기 전, **AI에게 우리가 기획한 서비스의 역할을 부여(Persona Prompting)**하여 가상으로 작동 과정을 시뮬레이션해 봅니다. 
이 단계를 통해 서비스가 유저에게 정말 유용한지 검증하고, 나중에 서비스 개발에 활용할 핵심 프롬프트를 미리 다듬을 수 있습니다.

**👉 우리가 결정한 아이디어 이름과 AI 에이전트의 구체적인 특징을 적어 코드 셀을 실행하세요.**

```python
# 3단계 프롬프트 생성 코드
service_name = "<플랜버디 (PlanBuddy)>" # 내가 선택한 AI 서비스 이름
ai_role_description = "<마감일에 찌든 학생들의 성향(벼락치기형, 계획형 등)을 파악한 뒤, 살짝의 유머와 격려를 섞어 멘탈을 케어해주며 1일 단위 맞춤형 할 일 목록을 짜주는 AI 스케줄러 멘토>" # AI의 역할과 성격 명시

prompt_step3 = f"""
우리가 해커톤에서 최종적으로 개발하기로 결정한 아이디어는 '{service_name}'야.
이 서비스의 핵심은 사용자와 친근하게 대화하며 실질적인 가치를 주는 AI 에이전트야.

지금부터 너는 이 AI 서비스의 실제 '프로토타입(시뮬레이터)' 역할을 맡아 수행해 줄 거야. 아래의 규칙을 엄격히 따라줘.

[AI 에이전트 페르소나 설정]
- 역할 및 성격: {ai_role_description}
- 대화 스타일: 청소년 유저가 친근함을 느끼도록 이모티콘을 적절히 섞은 친절한 말투 (또는 설정된 특별한 말투)

[시뮬레이션 규칙]
1. 내가 다음 대화에서 '시작'이라고 외치면, 너는 실제 사용자가 '{service_name}'에 처음 접속했을 때 보게 될 첫 환영 인사 메시지를 건네며 대화를 시작해 줘.
2. 실제 서비스 환경처럼 자연스럽게 질문을 던져 유저의 상황을 파악해 가며, 유저인 나와 최소 3번 이상 티키타카 대화를 주고받아야 해.
3. 대화 도중 AI의 페르소나를 잃지 마.

준비가 되었다면 절대 다른 설명은 하지 말고, 딱 한 줄로 "준비 완료! '시작'이라고 말해줘!"라고만 답변해 줘.
"""

print(prompt_step3.strip())
```

📌 **주의**: 3단계 프롬프트를 AI에게 입력하면 AI가 `준비 완료! '시작'이라고 말해줘!`라고 할 것입니다. 그때 여러분이 **"시작"**이라고 직접 채팅을 입력하여 가상 챗봇과 대화를 나누는 **'역할극 테스트'**를 진행해 보세요!

---
## 📋 [Step 4] 해커톤 아이디어 기획서 초안 작성하기

축하합니다! AI 코치와 깊은 대화를 나누며 아이디어가 매우 단단해졌을 것입니다.
이제 해커톤 당일 팀원들과 공유하거나 대회에 제출할 수 있는 **'한 페이지 기획서'** 형태로 지금까지의 대화 내용을 깔끔하게 요약해 달라고 AI에게 요청해 봅시다.

**👉 가상 테스트까지 마친 후, 아래 코드 셀을 실행하여 AI에게 마지막 명령을 전달하세요.**

```python
# 4단계 프롬프트 생성 코드
prompt_step4 = f"""
지금까지 너와 대화하며 내 AI 아이디어가 정말 구체적이고 실현 가능하게 발전했어! 고마워.
우리가 지금까지 나눈 모든 디자인 씽킹 대화 내용을 집약해서, 해커톤 제출용 'AI 아이디어 기획서' 초안을 마크다운 표 구조를 활용해 가독성 높게 작성해 줘.

[포함해야 할 항목]
1. 프로젝트명 (최종 확정된 세련된 이름)
2. 타겟 유저 및 페인 포인트 (누가 어떤 불편함을 겪고 있는가?)
3. 서비스 정의 (AI를 활용해 이를 어떻게 한 줄로 해결하는가?)
4. 핵심 AI 기능 (기능 2~3가지를 상세 기술)
5. 디자인 씽킹 청소년 자가진단 3요소:
   - 공감(Desirability): 왜 유저들이 이 서비스를 진심으로 원하고 좋아할까?
   - 기술적 실현 가능성(Feasibility): 해커톤 기간 내에 어떤 오픈소스나 LLM API를 조합하여 쉽게 구현할 수 있는가?
   - 유용성(Usability): 이 서비스가 지속된다면 청소년들의 삶에 어떤 긍정적 가치와 변화를 줄까?
6. 해커톤 개발 시작 전, 우리 팀이 멘토님께 조언을 구하거나 추가로 리서치해야 할 핵심 고려 사항 2가지
"""

print(prompt_step4.strip())
```
