````python
import os
from dotenv import find_dotenv, load_dotenv

# 프로젝트 루트의 .env 자동 탐색
load_dotenv(find_dotenv(usecwd=True), override=True)

APIM_BASE_URL   = os.environ["APIM_BASE_URL"].rstrip("/")
APIM_KEY        = os.environ["APIM_KEY"]
CHAT_MODEL      = os.getenv("CHAT_MODEL", "gpt-5.4")

print("✅ API 설정 완료!")
print(f"- APIM Base URL : {APIM_BASE_URL}")
print(f"- APIM Key      : {APIM_KEY[:6]}...{APIM_KEY[-4:]}")
print(f"- Chat model    : {CHAT_MODEL}")
````

### 🔧 AI 코치 호출 함수 준비

아래 셀은 이 워크북 전체에서 사용할 **AI 코치 호출 함수**를 정의합니다.  
한 번만 실행하면 됩니다!

````python
from openai import OpenAI
from IPython.display import display, Markdown

# OpenAI 클라이언트 생성 (Foundry Proxy 방식)
client = OpenAI(
    api_key="placeholder",                              # APIM은 헤더로 인증
    base_url=f"{APIM_BASE_URL}/{CHAT_MODEL}/",
    default_headers={"api-key": APIM_KEY},
)

def ask_coach(prompt, system_message=None, history=None, temperature=0.8):
    """
    AI 코치에게 질문하고 답변을 받는 함수.
    
    Parameters:
        prompt (str): 사용자의 질문/프롬프트
        system_message (str): AI의 역할 설정 (시스템 메시지)
        history (list): 이전 대화 내역 (멀티턴 대화 시 사용)
        temperature (float): 창의성 조절 (0.0~1.0, 높을수록 창의적)
    
    Returns:
        tuple: (AI 답변 텍스트, 업데이트된 대화 내역)
    """
    if history is None:
        history = []
    
    # 메시지 구성
    messages = []
    if system_message:
        messages.append({"role": "system", "content": system_message})
    messages += history
    messages.append({"role": "user", "content": prompt})
    
    # API 호출
    response = client.chat.completions.create(
        model=CHAT_MODEL,
        messages=messages,
        temperature=temperature,
    )
    
    answer = response.choices[0].message.content
    
    # 대화 내역 업데이트
    updated_history = history + [
        {"role": "user", "content": prompt},
        {"role": "assistant", "content": answer},
    ]
    
    return answer, updated_history

print("✅ AI 코치 함수 준비 완료! 이제 각 단계를 진행할 수 있습니다.")
````

## 📋 [Step 4] 해커톤 아이디어 기획서 초안 작성하기

축하합니다! AI 코치와 깊은 대화를 나누며 아이디어가 매우 단단해졌을 것입니다.
이제 해커톤 당일 팀원들과 공유하거나 대회에 제출할 수 있는 **'한 페이지 기획서'** 형태로 지금까지의 대화 내용을 깔끔하게 요약해 달라고 AI에게 요청해 봅시다.

**👉 아래 코드 셀을 실행하면 AI가 기획서 초안을 자동 생성합니다!**

````python
# 4단계: 기획서 초안 자동 생성
system_step4 = "너는 청소년 AI 해커톤의 전문 기획 코치야. 마크다운 형식으로 가독성 높게 기획서를 작성해 줘."

# 지금까지의 맥락을 종합하여 기획서 요청
prompt_step4 = f"""지금까지의 디자인 씽킹 과정을 정리해 줘.

[지금까지의 맥락]
- 관심 분야: {interest_area}
- 페인 포인트: {pain_point}
- 핵심 문제 정의(HMW): {chosen_hmw}
- 서비스 이름: {service_name}
- AI 에이전트 역할: {ai_role_description}

[요청사항]
우리가 지금까지 나눈 모든 디자인 씽킹 내용을 집약해서, 해커톤 제출용 'AI 아이디어 기획서' 초안을 마크다운 표 구조를 활용해 가독성 높게 작성해 줘.

[포함해야 할 항목]
1. 프로젝트명 (최종 확정된 세련된 이름)
2. 타겟 유저 및 페인 포인트 (누가 어떤 불편함을 겪고 있는가?)
3. 서비스 정의 (AI를 활용해 이를 어떻게 한 줄로 해결하는가?)
4. 핵심 AI 기능 (기능 2~3가지를 상세 기술)
5. 디자인 씽킹 청소년 자가진단 3요소:
   - 공감(Desirability): 왜 유저들이 이 서비스를 진심으로 원하고 좋아할까?
   - 기술적 실현 가능성(Feasibility): 해커톤 기간 내에 어떤 오픈소스나 LLM API를 조합하여 쉽게 구현할 수 있는가?
   - 유용성(Usability): 이 서비스가 지속된다면 청소년들의 삶에 어떤 긍정적 가치와 변화를 줄까?
6. 해커톤 개발 시작 전, 우리 팀이 멘토님께 조언을 구하거나 추가로 리서치해야 할 핵심 고려 사항 2가지
"""

answer_step4, history_step4 = ask_coach(prompt_step4, system_message=system_step4)

print("="*60)
print("📋 AI가 생성한 해커톤 아이디어 기획서 초안")
print("="*60)
display(Markdown(answer_step4))
````
